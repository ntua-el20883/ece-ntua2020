### 4. Εκπαίδευση μοντέλου με default τιμές στις παραμέτρους των ταξινομητών

Θα εκπαιδεύσουμε τους ακόλουθους ταξινομητές με το σύνολο εκπαίδευσης των δεδομένων μας:
1. Naive Bayes
2. KNeighborsClassifier
3. LogisticRegression
4. MLP με ένα κρυφό επίπεδο
5. SVC
6. Decision Tree
7. Random Forest


```python
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

 4.α. Εκπαιδεύστε (fit)  και τους  7 ταξινομητές που προαναφέρθηκαν (εμφανίστε τα δείγματα από το train set μαζί με την ετικέτα τους στην είσοδο του κάθε ταξινομητή), χρησιμοποιώντας τις default τιμές για όλες τις παραμέτρους τους.

```python
# Question 4a -- Train
print("Question 4a -- Train")
# Model 1: Naive Bayes
classifierNB = GaussianNB()
Naive_Bayes = Pipeline(steps=[('preprocessor', preprocessor), ('classifierNB', classifierNB)])
Naive_Bayes.fit(X_train, y_train)

# Model 2: K Nearest Neighbors
classifierKNN = KNeighborsClassifier(n_neighbors=5)
K_Nearest_Neighbors = Pipeline(steps=[('preprocessor', preprocessor), ('classifierKNN', classifierKNN)])
K_Nearest_Neighbors.fit(X_train, y_train)

# Model 3: Logistic Regresion
classifierLR = LogisticRegression()
Logistic_Regresion = Pipeline(steps=[('preprocessor', preprocessor), ('classifierLR', classifierLR)])
Logistic_Regresion.fit(X_train, y_train)

# Model 4: Multilayer Perceptron with one hidden level
classifierMLP = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000)
Multilayer_Perceptron = Pipeline(steps=[('preprocessor', preprocessor), ('classifierMLP', classifierMLP)])
Multilayer_Perceptron.fit(X_train, y_train)

# Model 5: Support Vector Classifier
classifierSVC = SVC()
Support_Vector_Classifier = Pipeline(steps=[('preprocessor', preprocessor),('classifierSVC', classifierSVC)])
Support_Vector_Classifier.fit(X_train, y_train)

# Model 6: Decision Tree
classifierDT = DecisionTreeClassifier()
Decision_Tree = Pipeline(steps=[('preprocessor', preprocessor),('classifierDT', classifierDT)])
Decision_Tree.fit(X_train, y_train)

# Model 7: Random Forest
classifierRF = RandomForestClassifier()
Random_Forest = Pipeline(steps=[('preprocessor', preprocessor),('classifierRF', classifierRF)])
Random_Forest.fit(X_train, y_train)

print("Training complete.")
```

4.β. Εφαρμόστε στα εκπαιδευμένα μοντέλα από το validation set μόνο τα δείγματα χωρίς την ετικέτα τους (predict).


```python
# Question 4b -- Predict
print("Question 4b -- Predict")
# Model 1: Naive Bayes
y_predNB = Naive_Bayes.predict(X_test)

# Model 2: K Nearest Neighbors
y_predKNN = K_Nearest_Neighbors.predict(X_test)

# Model 3: Logistic Regresion
y_predLR = Logistic_Regresion.predict(X_test)

# Model 4: Multilayer Perceptron with one hidden level
y_predMP = Multilayer_Perceptron.predict(X_test)

# Model 5: Support Vector Classifier
y_predSVC = Support_Vector_Classifier.predict(X_test)

# Model 6: Decision Tree
y_predDT = Decision_Tree.predict(X_test)

# Model 7: Random Forest
y_predRF = Random_Forest.predict(X_test)

print("Preditctions completed.")
```


4.γ. Συγκρίνετε την έξοδο του κάθε μοντέλου σε σχέση με τις αντίστοιχες ετικέτες του validation set και αξιολογήστε την επιδοσή τους χρησιμοποιώντας το F1 score.


```python
# Question 4c -- Evaluate
print("Question 4c -- Evaluate:")
# Model 1: Naive Bayes
print(classification_report(y_test, y_predNB))
F1_NB = f1_score(y_test, y_predNB, average='weighted', labels=np.unique(y_predNB))
print('F1 score for model Naive Bayes: ', F1_NB)
print("")

# Model 2: K Nearest Neighbors
print(classification_report(y_test, y_predKNN))
F1_KNN = f1_score(y_test, y_predKNN, average='weighted', labels=np.unique(y_predKNN))
print('F1 score for model K Nearest Neighbors: ', F1_KNN)
print("")

# Model 3: Logistic Regresion
print(classification_report(y_test, y_predLR))
F1_LR = f1_score(y_test, y_predLR, average='weighted', labels=np.unique(y_predLR))
print('F1 score for model Logistic Regresion: ', F1_LR)
print("")

# Model 4: Multilayer Perceptron with one hidden level
print(classification_report(y_test, y_predMP))
F1_MP = f1_score(y_test, y_predMP, average='weighted', labels=np.unique(y_predMP))
print('F1 score for model Multilayer Perceptron: ', F1_MP)
print("")

# Model 5: Support Vector Classifier
print(classification_report(y_test, y_predSVC))
F1_SVC = f1_score(y_test, y_predSVC, average='weighted', labels=np.unique(y_predSVC))
print('F1 score for model Support Vector Classifier: ', F1_SVC)
print("")

# Model 6: Decision Tree
print(classification_report(y_test, y_predDT))
F1_DT = f1_score(y_test, y_predDT, average='weighted', labels=np.unique(y_predDT))
print('F1 score for model Decision Tree: ', F1_DT)
print("")

# Model 7: Random Forrest
print(classification_report(y_test, y_predRF))
F1_RF = f1_score(y_test, y_predRF, average='weighted', labels=np.unique(y_predRF))
print('F1 score for model Random Forest: ', F1_RF)
print("")

print("Evaluations completed.")
```
